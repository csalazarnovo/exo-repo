```python
import numpy as np
```

```python
"n is a discrete variable between 1 and 365"
def directBeamAndTiltAngle(L):
    '''
    L is parameter for Latitude
    returns the direct beam variation for every day n throughout the year at location with Latitude L
    '''
    xValues = []
    dbValues = []
    taValues = []
    for i in range(1, 365):
        x = i
        A = 1160+75*np.sin((2*np.pi/365)*(i-275))
        k = 0.174+0.035*np.sin((2*np.pi/365)*(i-100))
        Delta = 23.45*np.sin((2*np.pi/365)*(i-81))
        Beta = 90-L+Delta
        m = 1/np.sin(Beta*2*np.pi/360)
        db = A*np.exp(-k*m)
        tilt = 90-Beta
        
        xValues.append(x)
        dbValues.append(db)
        taValues.append(tilt)
    return xValues, dbValues, taValues
```

```python
import pylab as plt
```

```python
def plotDirectBeam(LValues):
    plt.figure('Direct Beam')
    plt.clf()
    LValuesLabels = ['r', 'b', 'g', 'k', 'm']
    for i in range(len(LValues)):
        L = LValues[i]
        LLabel = LValuesLabels[i%len(LValuesLabels)]
        xValues, dbValues, taValues = directBeamAndTiltAngle(L)
        plt.plot(xValues, dbValues, LLabel,label = 'L:' + str(L))

LValues = (49, 35, 40, 45, 50)

plotDirectBeam(LValues)
```

```python
def plotTiltAngle(LValues):
    plt.figure('Tilt Angle')
    plt.clf()
    LValuesLabels = ['r', 'b', 'g', 'k', 'm']
    for i in range(len(LValues)):
        L = LValues[i]
        LLabel = LValuesLabels[i%len(LValuesLabels)]
        xValues, taValues, taValues = directBeamAndTiltAngle(L)
        plt.plot(xValues, taValues, LLabel,label = 'L:' + str(L))

LValues = (49, 35, 40, 45, 50)

plotTiltAngle(LValues)
```

```python
def getDirectBeam(L):
    xValues, dbValues, taValues = directBeamAndTiltAngle(L)
    print("Direct Beam on February 20 in Paris is " + str(round(dbValues[50],2)) + " W/m2")
    print("Tilt Angle on February 20 in Paris is " + str(round(taValues[50],2)) + "")
L = 49
getDirectBeam(L)
```
